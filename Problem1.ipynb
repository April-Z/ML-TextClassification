```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
```

```python
mu_H = 3.4
variance_H = 0.42
sigma_H = math.sqrt(variance_H)
x_H = np.linspace(mu_H - 3*sigma_H, mu_H + 3*sigma_H, 100)
```

```python
mu_N = 3
variance_N = 0.2433
sigma_N = math.sqrt(variance_N)
x_N = np.linspace(mu_N - 3*sigma_N, mu_N + 3*sigma_N, 100)
```

```python
PH = 1/3
PN = 2/3
```

```python
ax = plt.gca()
plt.plot(x_H, 0.5*PH*stats.norm.pdf(x_H, mu_H, sigma_H), color='red')
plt.plot(x_N, 0.5*PN*stats.norm.pdf(x_N, mu_N, sigma_N), color='blue')
plt.axvline(x=3.7236)
```

```python
plt.plot(x_H, (1/5)*PH*stats.norm.pdf(x_H, mu_H, sigma_H), color='orange')
plt.plot(x_N, (4/5)*PN*stats.norm.pdf(x_N, mu_N, sigma_N), color='purple')
plt.axvline(x=4.2379)
```

```python
ax = plt.gca()
plt.plot(x_H, 0.5*PH*stats.norm.pdf(x_H, mu_H, sigma_H), color='red')
plt.plot(x_N, 0.5*PN*stats.norm.pdf(x_N, mu_N, sigma_N), color='blue')
plt.axvline(x=3.7236)
plt.plot(x_H, (1/5)*PH*stats.norm.pdf(x_H, mu_H, sigma_H), color='orange')
plt.plot(x_N, (4/5)*PN*stats.norm.pdf(x_N, mu_N, sigma_N), color='purple')
plt.axvline(x=4.2379)
```
